```python
# Instalasi pustaka google-play-scraper
!pip install google-play-scraper

# Mengimpor pustaka yang diperlukan
from google_play_scraper import app, Sort, reviews_all
import pandas as pd
import csv

print("Persiapan selesai. Library berhasil dimuat.")
```

```python
# Mengambil SEMUA ulasan dari aplikasi Canva
scrapreview = reviews_all(
    'id.dana',                # ID aplikasi
    lang='id',                # Bahasa ulasan: Indonesia
    country='id',             # Negara: Indonesia
    sort=Sort.MOST_RELEVANT   # Urutan ulasan yang paling relevan
)

# Mengonversi hasil scraping menjadi Pandas DataFrame agar mudah dikelola di Python
df_dana = pd.DataFrame(scrapreview)

df_dana.to_csv('ulasan_dana.csv', index=False, encoding='utf-8')

print(f"Proses scraping selesai! Total {len(df_dana)} ulasan telah berhasil diambil.")
```
