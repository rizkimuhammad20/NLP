**BoW**

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

df = pd.read_csv('data_preprocessed.csv')

# Bag of Words
bow = CountVectorizer(max_features=500)
X_bow = bow.fit_transform(df['teks_processed'].fillna(''))

print('Shape BoW:', X_bow.shape)
print('\n10 kata pertama di vocabulary:')
print(bow.get_feature_names_out()[:10])

df_bow = pd.DataFrame(X_bow.toarray(), columns=bow.get_feature_names_out())
print('\nMatriks BoW (5 baris pertama):')
print(df_bow.head())
```

**TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=500)
X_tfidf = tfidf.fit_transform(df['teks_processed'].fillna(''))

print('Shape TF-IDF:', X_tfidf.shape)

df_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=tfidf.get_feature_names_out())
print('\nMatriks TF-IDF (5 baris pertama, dibulatkan):')
print(df_tfidf.head().round(3))
```

```python
# 10 kata dengan skor TF-IDF rata-rata tertingg
mean_tfidf = df_tfidf.mean().sort_values(ascending=False)
print('Top 10 kata dengan TF-IDF rata-rata tertinggi:')
print(mean_tfidf.head(10))
```
